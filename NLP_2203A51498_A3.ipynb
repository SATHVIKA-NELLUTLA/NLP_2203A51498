1.Implement Word Embeddings on following text)
    words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england'].

```python
import gensim
from gensim.models import Word2Vec

words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']

sentences = [words]

model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

for word in words:
    word_vector = model.wv[word]
    print(f"Vector representation of '{word}':\n{word_vector}\n")
```

2.Implement Word Embeddings using PCA (Principal Component Analysis) on following text)
    words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england'].

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

word_vectors = [model.wv[word] for word in words]

pca = PCA(n_components=2)
result = pca.fit_transform(word_vectors)

plt.figure(figsize=(8, 6))
plt.scatter(result[:, 0], result[:, 1])

for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]))

plt.show()
```

3.Execute Dependency Parsing on any text or on following text :
The quick brown fox jumps over the lazy dog

```python
import spacy

nlp = spacy.load('en_core_web_sm')

sentence = "The quick brown fox jumps over the lazy dog"

doc = nlp(sentence)

for token in doc:
    print(f"Word: {token.text} -> POS: {token.dep_} -> Dependency: {token.dep} Head: {token.head.text}")
```
